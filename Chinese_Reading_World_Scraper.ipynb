```python
import requests
import os
from urllib.error import HTTPError
import urllib
from bs4 import BeautifulSoup
from html.parser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.fed = []

    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
```

```python
def get_html_from_link(link, encoding):
    try:        
        url = urllib.request.urlopen(link).read()
        soup = BeautifulSoup(url.decode(encoding, 'replace'), "lxml")
        return soup
    except HTTPError as http_error:
        print("HTTPError for link:{0}".format(link))
        link += "l"
        url = urllib.request.urlopen(link).read()
        soup = BeautifulSoup(url.decode(encoding, 'replace'), "lxml")
        return soup

def write_to_output(filename, lines_to_wrte):
    if not os.path.exists(os.path.dirname(filename)):
        try:
            os.makedirs(os.path.dirname(filename))
        except OSError as exc: # Guard against race condition
            if exc.errno != errno.EEXIST:
                raise
    with open(filename, "w") as text_file:
        for line in lines_to_wrte:
            print(line, file=text_file)
```

```python
levels = ['beginning', 'intermediate', 'advanced']
num_units = 30
num_lessons = 10
for level in levels:
    for unit_num in range(num_units):
        unit_num += 1
        
        for lesson_num in range(num_lessons):
            lesson_num += 1
            unit_num_as_str = ""
            lesson_num_as_str = ""
            
            if(unit_num < 10):
                unit_num_as_str = '0' + str(unit_num)
            if(unit_num >= 10):
                unit_num_as_str = str(unit_num) 
            
            if lesson_num != 10:
                lesson_num_as_str = '0' + str(lesson_num)          
            if(lesson_num == 10):
                lesson_num_as_str = str(lesson_num) 

            link = 'http://collections.uiowa.edu/chinese/readings/{0}/{1}_audio/{2}_audio_u{3}/{4}_audio_u{5}_{6}.htm' \
                    .format(level, str(level[0]), str(level[0]), unit_num_as_str, str(level[0]), unit_num_as_str, lesson_num_as_str)
            
            encoding = 'gb2312'
            
            print(link)
            soup = get_html_from_link(link, encoding)
            
            text_in_p_elements_arr = soup.body.find_all('p')
            skill = str(text_in_p_elements_arr[2])
            title = str(text_in_p_elements_arr[3])
            text_in_story = text_in_p_elements_arr[4:-1]
            story = ""
            for text in text_in_story:
                story += str(text)
            
            skill = strip_tags(skill)
            title = strip_tags(title)
            story = strip_tags(story).replace("*", "")
            lines_to_write = [skill, title, story]
            
            filename = '%s/%s/%s.txt' % (level, unit_num, lesson_num)
            
            write_to_output(filename, lines_to_write)
            
            
```
